1) 머신러닝 문제 접근 방법

    - 전체 데이터 분석과 의사 결정 과정에서 머신러닝 알고리즘이 차지하는 부분은 보통 작다. 
    - 효과적으로 머신러닝을 이용하려면 넓은 시각으로 문제를 봐야한다.
    - 먼저 답을 얻고자 하는 질문이 어떤 종류인지 생각해야 한다.
    - 목표가 있다면 시스템 구축전 성공을 어떻게 정의하고 측정할지, 최종 솔루션이 비즈니스와 연구의 목표에 어떤 영향을 줄지 생각해야 한다. 
    - 예를 들어 부정거래 예측 모델을 개발하는 경우
        - 부정 거래 예측이 실제로 작동하는지 어떻게 측정?
        - 알고리즘 평가에 알맞은 데이터 보유?
        - 구축 후 솔루션이 비즈니스에 어떤 영향을 주는가?
        - 모델 효과가 비즈니스에 영향이 크지 않다면 분석할 가치가 떨어진다. 
    - 문제 정의와 영향과 성공을 평가할 방법을 안다면 데이터를 모으고 프로토 타입을 만드는 일이다. 
    - 모델 구축은 새로운 데이터를 모으고, 정제하고, 만들고, 평가하는 순환 사이클의 한 부분이다. 
    - 오류 분석시, 부족 데이터, 추가 수집 데이터, 작업 재구성에 대한 정보 획득 가능
    - 더 많은 데이터 수집, 작업 흐름을 바꾸는 것이 매개변수 튜닝을 위해 지속적인 그리드서치보다 이득일 수 있따. 
    
2) 프로토 타입에서 제품 까지 
    - 현재 학습한 머신러닝 라이브러리 들은 분석과 프로토타입을 빠르게 수행해준다. 
    - 다양한 기관에서 파이썬과 scikit-learn을 이용하지만, 복잡한 인프라를 가진 회사는 파이썬을 자사 시스템에 통합시키기
    
      어려울 수 있다. 
    - 이 경우 데이터 분석팀은 파이썬과R을 이용해 빠르게 아이디어를 시험하고 제품팀은 안정적이고 큰 규모의 
      
      시스템을 위해 Go, 스칼라, C++, Java같은 언어를 이용한다. 
    - 실제 분석을 서비스에 반영하려면 다양한 요구사항이 발생하여, 보통은 여러 언어를 사용해야 한다. 
    - 한 알고리즘이 큰 시스템에 정착되려면 신뢰성, 예측 가능성, 실행환경, 메모리 요구사항 같은 
      
      소프트웨어 공학적인 측면이 필요하다. 
    - 이런 환경에서 잘 작동하는 머신러닝 시스템 구축시 필요한 핵심 요소는 단순함이다. 
    - 각 단게에서 복잡도가 얼마나 커지는지, 각 요소가 데이터나 컴퓨터 인프라 변화에 얼마나 견고한지를 확인해야 한다. 
    - 구글 머신러닝 팀의 연구원들의 논문을 참고하면 대규모 장기 프로젝트에서 기술부채 이슈는 특히 중요하고 
    
      이를 통해 배운 경험이 작고 단기적인 시스템에서의 소프트웨어 구축에도 큰 도움이 될 것이다. 
      
3) 제품 시스템 테스트
    - 사전에 수집한 테스트 셋을 기초로 하여 알고리즘이 만든 예측을 평가하는 방법을 오프라인 평가라고 한다. 
    - 사용자에게 노출되는 머신러닝 시스템이라면 오프라인 평가는 첫 단계 일뿐이다. 
    - 다음은 전체 시스템에서 알고리즘이 적용된 이후에 평가하는 온라인 테스트 or 라이브 테스트이다. 
    - 실제 추천이나 검색결과를 변경하면 사용자의 행동을 크게 바꾸거나 예상치 못한 결과를 얻을 수 있다. 
    - 이러한 돌발상황을 방지하기 위해 일종의 블라인드 테스트인 A/B테스를 이용한다. ( A와 B알고리즘을 비교 노출하여 성공 지표 측정후 둘 중 하나 선택 )
      
      -> 보통 A는 새로운 모델이고 B는 기존 시스템이다.
    - A/B테스트 외에도 밴디트 알고리즘 같이 온라인 테스트를 위한 정교한 방법이 있다.
